### Description of Final Product

This infographic directs the reader through a new technique for finding unique music: statistical analysis. Beginning by introducing and explaining auditory features (numerical measurements of a song's qualities), it continues by showing how these features are related using a correlation heatmap. Then, the project progresses into an analysis of the features with the strongest correlation: acousticness and energy. Scatterplots, regplots, and outlier annotation are used to demonstrate how these features relate, and how they can be seperated and organized in a manner that highlights "outliers" songs that stand out from statistical norms of these acoustic measurements. Finally, the infrographic concludes with a more free-form interactive feature, displaying the information and links of the actual songs that are highlighted as outliers on the graph. It also describes the potential limitations of this analysis and this dataset.
